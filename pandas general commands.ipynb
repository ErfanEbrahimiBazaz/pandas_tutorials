````python
mask1 = (df_heart['age'] >= 40) & (df_heart['age'] < 60) 
mask2 = (df_heart['age'] >= 60) & (df_heart['age'] < 100) 

df_heart.loc[mask1, 'age_bin'] = 'B'
df_heart.loc[mask2, 'age_bin'] = 'C'

df_heart.head()[['age', 'age_bin']]
````
